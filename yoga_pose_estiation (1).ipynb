```python
#imorting library 
import cv2
import mediapipe as mp
import numpy as np
from matplotlib import pyplot
import glob
mp_drawing = mp.solutions.drawing_utils
mp_pose = mp.solutions.pose
```

```python
#changing the image to array of integer
path = "./input/yoga_images/"
filenames = [img for img in glob.glob(path+"/*.jpg")]

images = []
for img in filenames:
    n= cv2.imread(img)
    #cv2.imwrite('/input/yoga_images/', im_cv)
    images.append(n)

images = np.array(images)
```

```python
#Adding pose to the first 30 images
for i in range(30):
  with mp_pose.Pose(min_detection_confidence = 0.5, min_tracking_confidence = 0.5) as pose: 
    results = pose.process(images[i])
    #render detection:
    mp_drawing.draw_landmarks(images[i],results.pose_landmarks, mp_pose.POSE_CONNECTIONS)
```
